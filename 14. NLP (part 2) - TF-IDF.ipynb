TF-IDF 演算法包含了兩個部分：
- 詞頻（Term frequency，TF）</br>
> 詞頻指的是某一個分詞在該文件中出現的頻率，第 *t* 個詞出現在第 *d* 篇文件的頻率。</br>
> 例如，如果文件 1 總共有 2 個字，而第 blue 個字在文件 1 出現的次數是 1 次，因此 tf value = 1/2

- 逆向文件頻率（Inverse Document Frequency，IDF）</br>
> 常用字處理：例如 bright, sun 在 3 個文件中都出現，如此一來總體數字較少的文件的向量就會被這 2 個主導。但常用字對決定文件的分類沒有顯著的貢獻。

# 還原詞型 Lemmatization

https://easyai.tech/blog/all-you-need-to-know-about-text-preprocessing/

詞形還原（Lemmatization）是文本預處理中很重要的一部分，與詞幹提取（stemming）很相似。

簡單說來，詞形還原就是去掉單詞的詞綴，提取單詞的主幹部分，通常提取後的單詞會是字典中的「單詞」
不同於詞幹提取（stemming），提取後的單詞不一定會出現在單詞中。

例如： cars 還原為 car； ate 還原為 eat

<img src="https://miro.medium.com/max/1400/1*SCIQUg-UcQCAe0k-KZBUsQ.png" alt="drawing" width="800"/>


在Python 的 nltk模塊中，使用WordNet為我們提供了穩健的詞形還原的函數。如以下示例Python代碼：

``` python 

from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
# lemmatize nouns
print(wnl.lemmatize('cars', 'n'))
print(wnl.lemmatize('men', 'n'))
```
> car </br>
> men

``` python
# lemmatize verbs
print(wnl.lemmatize('running', 'v'))

```
> run </br>
> eat

``` python
# lemmatize adjectives
print(wnl.lemmatize('saddest', 'a'))

```
> sad </br>
> fancy

````python
#@title function
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')


# 获取单词的词性
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        # return None
        return wordnet.NOUN


# 還原詞性，
def normalize_text(text):
    result = []

    for i in nltk.word_tokenize(text.lower()) :

        pos_tag = nltk.pos_tag([i])
        lemm_words = nltk.WordNetLemmatizer().lemmatize( pos_tag[0][0], get_wordnet_pos( pos_tag[0][1]) )
        
        result.append(lemm_words)

    return [x.lower() for x in result]


print(normalize_text('community communities give given gave courses updated rotating varying boxes europe european cook cooker logistic	logistics'))

````

````python
import re

def text_cut(text):

    cut_words = jieba.lcut(text.lower())

    valid_text_split = []
    for word in cut_words:
        # print(word)
        # remove stopwords
        if ( word not in cn_stopwords and word not in en_stopwords and re.sub(r'[^\w]', '', word) != '') or word in remove_from_stopword:
             valid_text_split.append(word)

    # return ' '.join(filter(str.isalnum, valid_text_split))
    # return ','.join( valid_text_split)
    return [x for x in valid_text_split ]

# print(text)
# text_cut(text)

sf_text = '7-11便利店取件 SF Express - 7-11 Store Pickup'
print(jieba.lcut(sf_text))

text_cut(sf_text)
````

````python
# 文字前處理，(1)還原詞性 (2)斷詞 (3)移除中文＆英文的 stop words (4)移除特殊符號

def word_cut_to_doc( text ) : 
 
    # 針對商品名稱分詞
    cut_words = text_cut(text)
    cut_words = ','.join( cut_words)

    # 還原英文詞型態
    content_normalize = normalize_text(cut_words)

    # 將中文數字轉成阿拉伯數字， 例如：全自動十骨摺疊傘 >> 全自動10骨摺疊傘     
    content = [cn2an.transform(x, "cn2an") for x in content_normalize ]

    # 繁體中文轉成簡體 + 剔除含有數字的文字，例如: 三件 五折 
    result = [cc.convert(x) for x in content if containsNumber(x) is False]

    # 剔除只有一個 digit 的詞
    result = [x for x in result if len(x) > 1 ]

    # output 格式為 List
    return result

print( word_cut_to_doc( '风火轮任天堂游戏角色合金小车*]＊ 一组八入 店  色' ) )
print( word_cut_to_doc('711 便利店 取件 sf express 711 store pickup') )
````

````python
#@title *function pre-process words
# 文字前處理，(1)還原詞性 (2)斷詞 (3)移除中文＆英文的 stop words (4)移除特殊符號

def word_cut_to_doc( text ) : 
 
    # 針對商品名稱分詞
    cut_words = text_cut(text)
    cut_words = ' '.join( cut_words)

    # 還原英文詞型態
    content_normalize = normalize_text(cut_words)

    # 繁體中文轉成簡體 
    result = [cc.convert(x) for x in content_normalize ]

    # output 格式為 List
    return result

print( word_cut_to_doc( '风火轮任天堂游戏角色合金小车*]＊ 一组八入 ' ) )
print( word_cut_to_doc('7-11 便利店 取件 sf express 7-11 store pickup') )
````

**TF-IDF 演算法**

<img src="https://1.bp.blogspot.com/-tnzPA6dDtTU/Vw6EWm_PjCI/AAAAAAABDwI/JatHtUJb4fsce9E-Ns5t02_nakFtGrsugCLcB/s1600/%25E8%259E%25A2%25E5%25B9%2595%25E5%25BF%25AB%25E7%2585%25A7%2B2016-04-14%2B%25E4%25B8%258A%25E5%258D%25881.39.07.png" alt="drawing" width="800"/>




TF-IDF 演算法包含了兩個部分：
- 詞頻（Term frequency，TF）</br>
> 詞頻指的是某一個分詞在該文件中出現的頻率，第 *t* 個詞出現在第 *d* 篇文件的頻率。</br>
> 例如，如果文件 1 總共有100個字，而第 1 個字在文件 1 出現的次數是12次，因此 tf value = 12/100
- 逆向文件頻率（Inverse Document Frequency，IDF）</br>
> 常用字處理：比如說， the 在每個文件中都出現好多次，如此一來總體數字較少的文件的向量就會被 the 這個字所主導，但 the 這個字其實沒什麼特別的意義。


<img src="https://taweihuang.hpd.io/wp-content/uploads/2017/03/%E5%9C%96%E7%89%871-1.png" alt="drawing" width="800"/>




````python
#@title *build model { form-width: "150px"}

def semblance(text, corpus):

    content = word_cut_to_doc( text )

    # test data 的 bag_of_word
    doc_text_vec = dictionary.doc2bow(content)

    # training data 裡，每個文件中，每個分詞的 tfidf 值
    tfidf = models.TfidfModel(corpus)
    # print('tfidf',tfidf)

    # 对每个文档，分析要测试文档的相似度
    index = similarities.SparseMatrixSimilarity( tfidf[corpus], num_features=len(dictionary.keys()) )
    # print('index', index)

    sim = index[tfidf[doc_text_vec]]
    # print('sim', sim)

    sim_sorted = sorted(enumerate(sim), key=lambda x: -x[1])
    return sim_sorted[ : 10]
````

Cosine Similarity (餘弦相似度) 是在計算文本相似度時相當常見的一種計算方法，原理也相當易懂，基本上就是計算『兩向量』之間的 Cosine 夾角。

- 夾角越大，代表兩個向量越「不像」
- 夾角越小，代表兩個向量越「相似」


<img src="https://i0.wp.com/clay-atlas.com/wp-content/uploads/2020/03/image-72.png?w=734&ssl=1" alt="drawing" width="500"/>


```python

from sklearn.metrics.pairwise import cosine_similarity


# Vectors
vec_a = [1, 2, 3, 4, 5]
vec_b = [1, 3, 5, 7, 9]

# Dot and norm
dot = sum(a*b for a, b in zip(vec_a, vec_b))
norm_a = sum(a*a for a in vec_a) ** 0.5
norm_b = sum(b*b for b in vec_b) ** 0.5

# Cosine similarity
cos_sim = dot / (norm_a*norm_b)

# Results
print('My version:', cos_sim)
print('Scikit-Learn:', cosine_similarity([vec_a], [vec_b]))

```

```python
My version: 0.9972413740548081
Scikit-Learn: [[0.99724137]]
```

````python
#@title *build model { form-width: "150px"}

def semblance_for_train(text, corpus):

    # content = word_cut_to_doc( text )

    # test data 的 bag_of_word
    doc_text_vec = dictionary.doc2bow(text)

    # training data 裡，每個文件中，每個分詞的 tfidf 值
    tfidf = models.TfidfModel(corpus)
    # print('tfidf',tfidf)

    # 对每个文档，分析要测试文档的相似度
    index = similarities.SparseMatrixSimilarity( tfidf[corpus], num_features=len(dictionary.keys()) )
    # print('index', index)

    sim = index[tfidf[doc_text_vec]]
    # print('sim', sim)

    sim_sorted = sorted(enumerate(sim), key=lambda x: -x[1])
    return sim_sorted[ : 10]
````

````python
import statistics 
import pandas as pd

df = df_doc.copy().head(10)


class_result = []

for i in range(df.shape[0] ):

    handle = df['handle'][i]
    ori_type = df['delivery_type'][i]
    text = df['doc'][i]

    outcomes = semblance_for_train(text, corpus)

    print(handle, ' : outcomes',outcomes[:6])

    source_platform = []
    for i in outcomes[1:10] :
        answer_index = i[0]
        weight = i[1]
        delivery_platform = df_doc.iloc[ answer_index , df_doc.columns.get_loc("delivery_type") ]

        source_platform.append([weight, delivery_platform, answer_index])

    df_temp = pd.DataFrame(source_platform, columns = ['weight','delivery_type', 'answer_index'])

    weighted_tag = df_temp.groupby(['delivery_type']).agg({'weight': 'sum', 'answer_index':'count'}).reset_index().sort_values(by='weight',ascending = False).reset_index(drop=True)
    new_platform = weighted_tag.iloc[0, weighted_tag.columns.get_loc("delivery_type") ]
    
    similarity = round( weighted_tag.iloc[0, weighted_tag.columns.get_loc("weight") ]  ,4)
    count = weighted_tag.iloc[0, weighted_tag.columns.get_loc("answer_index")]

    class_result.append([handle, new_platform, similarity, count ,ori_type, text])
    print(handle, new_platform, similarity, count ,ori_type, text)

# class_result
pd.DataFrame(class_result, columns = ['handle','new_type','similarity','count','ori_type', 'text']).sort_values(by= 'similarity')
````

````python
# Create the Document-Word matrix
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

df_vec = df_doc.copy()
df_vec.head()
vectorizer = CountVectorizer(analyzer='word',       
                             min_df=5,              
                             lowercase=True,                   # convert all words to lowercase
                            #  stop_words = contain_number,
                            #  token_pattern='[a-zA-Z0-9]{3,}',  # num chars > 3
                            #  max_features=50000,             # max number of uniq words
                            )

data_vectorized = vectorizer.fit_transform( [','.join(map(str, l)) for l in df_vec['doc']]  )

vectorizer
````

````python
#@title Check Sparsity
# 稀疏性值越大越好，代表該文字有意義

# Materialize the sparse data
data_dense = data_vectorized.todense()

# Compute Sparsity = Percentage of Non-Zero cells
print("Sparsity: ", ((data_dense > 0).sum()/data_dense.size)*100, "%")
````

````python
# Build LDA model with sklearn
lda_model = LatentDirichletAllocation(n_components = 15,            # Number of topics
                                      max_iter=10,               # Max learning iterations
                                      learning_method='batch',   
                                      random_state=100,          # Random state
                                     )
lda_output = lda_model.fit_transform(data_vectorized)

print(lda_model) 
````

````python
# How to see the dominant topic in each document?
# Create Document - Topic Matrix
lda_output = lda_model.transform(data_vectorized)

# column names
topicnames = ["Topic" + str(i) for i in range(lda_model.n_components)]

# index names
data = df_vec['doc']
docnames = ["Doc" + str(i) for i in range(len(data))]

# Make the pandas dataframe
df_document_topic = pd.DataFrame(np.round(lda_output, 2), columns=topicnames, index=docnames)

# Get dominant topic for each document
dominant_topic = np.argmax(df_document_topic.values, axis=1)
df_document_topic['dominant_topic'] = dominant_topic

# Styling
def color_green(val):
    color = 'green' if val > .1 else 'black'
    return 'color: {col}'.format(col=color)

def make_bold(val):
    weight = 700 if val > .1 else 400
    return 'font-weight: {weight}'.format(weight=weight)

# Apply Style
df_document_topics = df_document_topic.style.applymap(color_green).applymap(make_bold)
df_document_topics
````

````python
# Review topics distribution across documents
df_topic_distribution = df_document_topic['dominant_topic'].value_counts().reset_index(name="Num Documents")
df_topic_distribution.columns = ['Topic Num', 'Num Documents']
df_topic_distribution
````

````python
# the Topic’s keywords
# Topic-Keyword Matrix
df_topic_keywords = pd.DataFrame(lda_model.components_)

# Assign Column and Index
df_topic_keywords.columns = vectorizer.get_feature_names()
df_topic_keywords.index = topicnames

# View
df_topic_keywords.head()
````

````python
# Get the top 15 keywords each topic
# Show top n keywords for each topic
def show_topics(vectorizer=vectorizer, lda_model=lda_model, n_words=20):
    keywords = np.array(vectorizer.get_feature_names())
    topic_keywords = []
    for topic_weights in lda_model.components_:
        top_keyword_locs = (-topic_weights).argsort()[:n_words]
        topic_keywords.append(keywords.take(top_keyword_locs))
    return topic_keywords

topic_keywords = show_topics(vectorizer=vectorizer, lda_model=lda_model, n_words=20)        

# Topic - Keywords Dataframe
df_topic_keywords = pd.DataFrame(topic_keywords)
df_topic_keywords.columns = ['Word '+str(i) for i in range(df_topic_keywords.shape[1])]
df_topic_keywords.index = ['Topic '+str(i) for i in range(df_topic_keywords.shape[0])]
df_topic_keywords
````
