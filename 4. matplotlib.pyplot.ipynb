````
#@title Basic plot

import matplotlib.pyplot as plt

#1) plot a figure which default size = (6.4, 4.8)
plt.figure()

#2) Feed data and decide the chart type 
x = [0, 0.6 ,1]
y = [0, 0.3 , 1]
plt.plot(x, y)

#3) Finetune figure elements
plt.xlim([0.0, 0.5])
plt.ylim([0.0, 1.0])
plt.title('Test figure')  

#4) Show figure (Especially on Editor)
# plt.show()
````

````
#@title Figure elements
import matplotlib.pyplot as plt

# Set up figure size
plt.figure (figsize = (6, 4.5))              

# Plot bar chart in the figure
plt.bar(x=['A','B','C'], height=[3.1,7,4.2], color=['red','purple','g'])

plt.title('Bar Chart Demo', fontsize = 18)
plt.xlabel('r (m)', fontsize = 14)                        # 設定坐標軸標籤
plt.xticks(fontsize = 12)                                 # 設定 x軸標示 字級大小
plt.yticks(fontsize = 12)                                 # 設定 y軸標示 字級大小
plt.grid(color = 'red', linestyle = '-', linewidth = 1)  # 設定格線顏色、樣式、寬度
plt.ylim(0, 20)                                           # 設定 y軸 數字範圍

plt.show()
````

````
#@title Basic Stake Line Chart

import numpy as np
import matplotlib.pyplot as plt



rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(12, 5))

ax.stackplot(yrs, WITH orders_hourly AS (
SELECT 
  *
FROM (    
    SELECT 
      _id,
      seller_id,
      ROW_NUMBER() OVER(PARTITION BY _id ORDER BY updated_at DESC) rn
    FROM 
      `shopline-142003.datawarehouse.orders_hourly`
    WHERE 
      created_at >= '2020-01-01'
      AND created_at < '2020-07-01'
    )
WHERE 
  rn = 1
)
, order_deliveries_hourly AS (
SELECT 
  seller_id,
  order_id,
  total_currency_iso,
  status,
  delivery_status,
  ref_shipment_id,
  MAX( LENGTH(ref_shipment_id) ) OVER(PARTITION BY order_id) AS count_shipment_id , 
  created_at,

  shipped_at,  
  arrived_at,
  collected_at,
  updated_at,
  COUNT( DISTINCT shipped_at ) OVER (PARTITION BY order_id ) AS count_shipped_at
FROM 
  `shopline-142003.datawarehouse.order_deliveries_hourly` AS order_deliveries_hourly
  LEFT JOIN orders_hourly ON orders_hourly._id = order_deliveries_hourly.order_id
WHERE 
  created_at >= '2020-01-01'
  AND created_at < '2020-07-01'
  AND return_order_id IS NULL
)
SELECT 
  seller_id,
  COUNT( DISTINCT IF(count_shipped_at > 1  AND count_shipment_id = 0, order_id, NULL) ) multiple_delivered,
  COUNT( DISTINCT order_id) AS nOrders,
  ROUND( COUNT( DISTINCT IF(count_shipped_at > 1  AND count_shipment_id = 0, order_id, NULL) )  / COUNT( DISTINCT order_id) * 100, 2) AS rate 
FROM 
  order_deliveries_hourly
GROUP BY 
  seller_id
HAVING 
  rate > 0
ORDER BY 
  rate DESC, labels=['Eastasia', 'Eurasia', 'Oceania'])

ax.set_title('Combined debt growth over time', fontsize = '20')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim( xmin=yrs[0], xmax=yrs[-1] )
````

# Pie & Donut Chart

````
#@title Basic Pie & Chart
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.2, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5))

# 1) Pie Chart
# https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/pie_features.html

ax1.pie(sizes, 
       explode = explode, 
       labels = labels, 
       autopct ='%1.2f%%',
       shadow = True, 
       startangle = 90)

#2) Donut Chart
# https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/nested_pie.html?highlight=pie%20chart
ax2.pie(sizes,
        labels = labels, 
        autopct ='%1.f%%',
        wedgeprops=dict(width= 0.3, edgecolor='w'), # Donut Chart 
        )

ax2.set_title("Donut Chart")

fig.suptitle('Pie & Donut', fontsize=16)


plt.show()
````

````
#@title Load Data

# Provide Credentials
from google.colab import auth
auth.authenticate_user()
print('Authenticated')

# Run the query
from datetime import datetime
from google.cloud import bigquery

client = bigquery.Client( 'shopline-142003')

print( 'Query Start : ', datetime.now())

row_data = client.query('''

WITH pos_trial AS (
  SELECT 
    owner_id AS merchant_id
  FROM (
    SELECT 
      owner_id, 
      target_key, 
      fee_cents, 
      payment_data_payment_gateway, 
      ROW_NUMBER() OVER(PARTITION BY owner_id ORDER BY created_at DESC, updated_at DESC) AS row_number
    FROM 
      `shopline-142003.datawarehouse.subscriptions_daily`
    ) 
  WHERE 
    row_number = 1 AND 
    REGEXP_CONTAINS(target_key, 'pos') AND 
    (fee_cents = 0 OR payment_data_payment_gateway = "free")
    
)
,subscriptions_daily AS (
  SELECT 
    DISTINCT owner_id 
  FROM (
    SELECT
      owner_id,
      count( IF( NOT REGEXP_CONTAINS(target_key, r"trial"), 1, NULL)) OVER (PARTITION BY  owner_id ORDER BY created_at ASC) AS no_of_pay_plan
    FROM
      `shopline-142003.datawarehouse.subscriptions_daily`  
    )
  WHERE 
    no_of_pay_plan > 0
)
, merchants_hourly AS (
  SELECT 
      _id AS merchant_id,
      handle,
      plan_key
  FROM (
    SELECT 
      _id, 
      handle,
      email,
      CASE 
        WHEN REGEXP_CONTAINS(current_plan_key, r"locked") THEN "Locked"
        WHEN REGEXP_CONTAINS(current_plan_key, r"free|_trial") OR _id IN (SELECT owner_id FROM pos_trial) THEN "Trial"
        WHEN REGEXP_CONTAINS(current_plan_key, r"o2o") THEN "O2O"
        WHEN REGEXP_CONTAINS(current_plan_key, r"enterprise_year") THEN "Enterprise"
        WHEN REGEXP_CONTAINS(current_plan_key, r"crossborder_year") THEN "CrossBorder"
        WHEN REGEXP_CONTAINS(current_plan_key, r"premium_") THEN "Premium"
        WHEN REGEXP_CONTAINS(current_plan_key, r"business_*") THEN "Business"
        WHEN REGEXP_CONTAINS(current_plan_key, r"advanced_*") THEN "Advanced"
        WHEN REGEXP_CONTAINS(current_plan_key, r"pos_") THEN "POS"
        WHEN REGEXP_CONTAINS(current_plan_key, r"pro") THEN "PRO"
        WHEN REGEXP_CONTAINS(current_plan_key, r"standard") THEN "Standard"
        ELSE "Basic" 
        END as plan_key,
      ROW_NUMBER() OVER (PARTITION BY _id ORDER BY updated_at DESC) row_number
    FROM 
      `shopline-142003.datawarehouse.merchants_hourly` 
    )
  WHERE
      row_number = 1
      -- Exclude merchant email with "shoplineapp.com" or "shopline.com"
      AND NOT REGEXP_CONTAINS(email, r"@shopline\w*.com") 
      AND _id IN (SELECT owner_id FROM subscriptions_daily )
)
, products_hourly AS (
  SELECT 
    _id AS product_id, 
    owner_id AS merchant_id,
    status,
    type,
    product_type
  FROM (
    SELECT 
      _id, 
      owner_id, 
      status, 
      type,
      product_type,
      ROW_NUMBER() OVER(PARTITION BY _id ORDER BY updated_at DESC) row_number
    FROM 
      `shopline-142003.datawarehouse.products_hourly`
    )
  WHERE
    row_number = 1
    AND status IN ("active", "draft", 'hidden')
)
SELECT 
  products_hourly.merchant_id,
  plan_key,
  handle,
  COUNT( DISTINCT IF(products_hourly.type = 'product' OR products_hourly.product_type = 'product', product_id, NULL) ) AS nProducts,
  COUNT( DISTINCT IF( (products_hourly.type = 'product' OR products_hourly.product_type = 'product') AND status = 'active', product_id, NULL) ) AS nProducts_Publishing
FROM 
  products_hourly
  JOIN merchants_hourly ON merchants_hourly.merchant_id = products_hourly.merchant_id
GROUP BY 
  plan_key,
  handle,
  products_hourly.merchant_id
HAVING 
  nProducts > 0
ORDER BY 
  nProducts DESC

''' ).to_dataframe()

print( 'Query Done : ', datetime.now())

print('shape = ', row_data.shape)
row_data.sample(5)
````

````
#@title Load Data

#@title Provide Credentials
from google.colab import auth
auth.authenticate_user()
print('Authenticated')

#@title Query data from Bigquery : df_staff
from datetime import datetime
from google.cloud import bigquery
client = bigquery.Client( 'shopline-142003')

print( 'Start : ', datetime.now())

df_staff = client.query('''

WITH pos_trial AS (
  SELECT 
    owner_id AS merchant_id
  FROM (
    SELECT 
      owner_id, 
      target_key, 
      fee_cents, 
      payment_data_payment_gateway, 
      ROW_NUMBER() OVER(PARTITION BY owner_id ORDER BY created_at DESC, updated_at DESC) AS row_number
    FROM 
      `shopline-142003.datawarehouse.subscriptions_daily`
    ) 
  WHERE 
    row_number = 1 AND 
    REGEXP_CONTAINS(target_key, 'pos') AND 
    (fee_cents = 0 OR payment_data_payment_gateway = "free")
)
, subscriptions_daily AS (
  SELECT 
    DISTINCT owner_id 
  FROM (
    SELECT
      owner_id,
      count( IF( NOT REGEXP_CONTAINS(target_key, r"trial"), 1, NULL)) OVER (PARTITION BY  owner_id ORDER BY created_at ASC) AS no_of_pay_plan
    FROM
      `shopline-142003.datawarehouse.subscriptions_daily`  
    )
  WHERE 
    no_of_pay_plan > 0
)
, merchants_hourly AS (
  SELECT 
    _id,
    handle,
    current_plan_key,
    email,
    CASE 
        WHEN REGEXP_CONTAINS(current_plan_key, r"locked") THEN "Locked"
        WHEN REGEXP_CONTAINS(current_plan_key, r"free|_trial") OR _id IN (SELECT owner_id FROM pos_trial) THEN "Trial"
        WHEN REGEXP_CONTAINS(current_plan_key, r"o2o") THEN "O2O"
        WHEN REGEXP_CONTAINS(current_plan_key, r"enterprise_year") THEN "Enterprise"
        WHEN REGEXP_CONTAINS(current_plan_key, r"crossborder_year") THEN "CrossBorder"
        WHEN REGEXP_CONTAINS(current_plan_key, r"premium_") THEN "Premium"
        WHEN REGEXP_CONTAINS(current_plan_key, r"business_*") THEN "Business"
        WHEN REGEXP_CONTAINS(current_plan_key, r"advanced_*") THEN "Advanced"
        WHEN REGEXP_CONTAINS(current_plan_key, r"pos_") THEN "POS"
        WHEN REGEXP_CONTAINS(current_plan_key, r"pro") THEN "PRO"
        WHEN REGEXP_CONTAINS(current_plan_key, r"standard") THEN "Standard"
        ELSE "Basic" 
        END as plan_key,
    ROW_NUMBER() OVER (PARTITION BY _id ORDER BY updated_at DESC) row_number
  FROM 
    `shopline-142003.datawarehouse.merchants_hourly`
  WHERE 
    _id NOT IN ('5bf76f982cf7f2000142280a')
)
, abilities_hourly AS (
  SELECT 
    *,
    ROW_NUMBER() OVER (PARTITION BY owner_id ORDER BY updated_at DESC) row_number
  FROM
    `shopline-142003.datawarehouse.abilities_hourly`
)
, users_hourly AS (
SELECT 
  _id,
  email,
  sign_in_count,
  is_staff_of_merchant_ids,
  COUNT(IF(owner_id = '536148ab1311667d72000001', 1, NULL)) OVER (PARTITION BY _id ORDER BY updated_at DESC) -1 admin_sign_in_count,
  ROW_NUMBER() OVER (PARTITION BY _id ORDER BY updated_at DESC) row_number
FROM 
  `shopline-142003.datawarehouse.users_hourly`
)
SELECT
  merchants_hourly._id AS merchant_id,
  handle,
  plan_key,
  COUNT( DISTINCT abilities_hourly.owner_id) nAssignedStaffs,
  COUNT( DISTINCT IF( admin_sign_in_count > 0, abilities_hourly.owner_id, NULL)) nSigned_in_Staffs
FROM 
  abilities_hourly
  JOIN merchants_hourly 
    ON merchants_hourly.row_number = 1
    AND merchants_hourly._id = abilities_hourly.resource_owner_id
    AND NOT REGEXP_CONTAINS(merchants_hourly.email, r"@shopline\w*.com")
  JOIN users_hourly
    ON LENGTH(users_hourly.is_staff_of_merchant_ids) > 0
    AND LENGTH(users_hourly.is_staff_of_merchant_ids) > LENGTH( Replace(users_hourly.is_staff_of_merchant_ids, resource_owner_id, ''))
    AND users_hourly._id = abilities_hourly.owner_id
    
WHERE 
  abilities_hourly.row_number = 1
  AND abilities_hourly.status != 'removed'
  AND merchants_hourly._id IN (SELECT owner_id FROM subscriptions_daily )
GROUP BY
  merchants_hourly._id,
  handle,
  plan_key

HAVING 
  nAssignedStaffs > 0
ORDER BY
  nAssignedStaffs DESC

''' ).to_dataframe()

print( 'Done : ', datetime.now())
df_staff.sample(10)
````

````
#@title Visualization

# Referenced from 
# https://stackoverflow.com/questions/33927591/add-either-a-density-or-box-plot-to-the-margins-of-a-plot-in-matplotlib

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D

df = df_staff.copy()
plans = df.groupby(by = ['plan_key']).agg({'merchant_id': 'nunique'}).reset_index()

colors = ('red', 'yellow','blue', 'pink', 'orange', 'gray', 'purple', 'Navy','aqua','green','black' )

left = 0.1
bottom = 0.1
top = 0.9
right = 0.9

fig = plt.figure(figsize=(10, 10))
main_ax = plt.axes([left,bottom,right-left,top-bottom])

# create axes to the top and right of the main axes and hide them
top_ax = plt.axes([left, top, right - left, 1-top])
plt.axis('off')
right_ax = plt.axes([right,bottom,1-right,top-bottom])
plt.axis('off')

legend_elements = []

for i in range(len(plans)):
    x = df[df['plan_key'] == plans.iloc[i,0]].nAssignedStaffs
    y = df[df['plan_key'] == plans.iloc[i,0]].nSigned_in_Staffs
    main_ax.plot(x,  y, 'o' , color=colors[i%len(colors)],  alpha=0.3, markersize=10)
    legend_elements.append( Line2D([0], [0], marker='o', markerfacecolor=colors[i%len(colors)], markersize=10,  alpha=0.5, lw=0, label=plans.iloc[i,0] ))


x = df['nAssignedStaffs'] 
top_ax.boxplot(x, positions=[0] , meanline = True, vert=False, widths=1.)
top_ax.set_title('No. of Staff Distribution', y= 0 , fontsize= 10)

# label 90 percentile on box chart
x_avg = x.quantile(0.90)
top_ax.axvline(x_avg, color='k', linestyle='dashed', linewidth=1)
top_ax.text(1, 1.5, '90 percentile: {:.0f}'.format(x_avg))

# label 50 percentile on box chart
x_avg = x.quantile(0.50)
top_ax.axvline(x_avg, color='k', linestyle='dashed', linewidth=1)
top_ax.text(1, 1.2, '50 percentile: {:.0f}'.format(x_avg))


# set annotations on the chart
main_ax.set_title('Staff Feature Usage', {'fontsize': 14, 'fontweight':"bold"}, y=1.1)
main_ax.set_ylabel('No. of Logged-in Staffs') 
main_ax.set_xlabel('No. of Staffs') 

# set the limits to the box axes
top_ax.set_xlim(main_ax.get_xlim())
top_ax.set_ylim(-1,1)
right_ax.set_ylim(main_ax.get_ylim())
right_ax.set_xlim(-1,1)

main_ax.legend(handles=legend_elements, loc='lower right')

plt.show()
````
