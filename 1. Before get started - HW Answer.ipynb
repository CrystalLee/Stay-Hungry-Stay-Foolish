# Q3 Is there anyone visited the same pages

````
shoplytics_usage = {"A":["Customer","Overview","Orders"], 
                    "B":["Customer","Overview","Orders", "Product"], 
                    "C":["Overview","Customer","Orders"], 
                    "D":["Orders","Overview","Customer"], 
                    "E":["Customer","Overview","Product","Orders"]}

temp={}

for key, value in shoplytics_usage.items():
    #Because the page order is different between users, need to tidy up the orders
    value.sort() 
    temp[key]=tuple(value) 
    # We want to use "Pages" as key for the next loop, 
    # however list can't be used as key in dictionary, so change the value type into tuple

answer = {}
for key, value in temp.items():
    if value not in answer:
        answer[value] = [key]
    else:
        answer[value].append(key)

answer
# final.values()
````

````
import itertools
import pprint

all_page = set(itertools.chain(*shoplytics_usage.values()))
page_size, search_min_size = len(all_page), 3

result = {}
for num in range(page_size - search_min_size+1):
    for subset in itertools.combinations(all_page, page_size-num):
        result[subset] = [user for user, pages in shoplytics_usage.items() if set(subset).issubset(pages)]

pprint.pprint(result)
````
