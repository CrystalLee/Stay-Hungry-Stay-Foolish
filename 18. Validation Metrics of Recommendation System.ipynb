# What algorithms are used?


## 1.Personalised Video Ranking (PVR) 
This algorithm is a general-purpose one, which usually filters down the catalog by a certain criteria (e.g. Violent TV Programmes, US TV shows, Romance, etc), combined with side features including user features and popularity.


<img src="https://miro.medium.com/max/1400/1*IeYlZKqQy8vwh5UaWbq0JQ.png" alt="drawing" width="1000"/>


## 2.Top-N Video Ranker
Similar to PVR except that it only looks at the head of the rankings and looks at the entire catalog. It is optimised using metrics that look at the head of the catalog rankings (e.g. MAP@K, NDCG).

<img src="https://miro.medium.com/max/1400/1*5MnyClVj3e7BiCe97k5KNQ.png" alt="drawing" width="1000"/>




## 3.Trending Now Ranker 
This algorithm captures temporal trends which Netflix deduces to be strong predictors. These short-term trends can range from a few minutes to a few days. These events/trends are typically:
1. Events that have a seasonal trend and repeat themselves (e.g. Valentines day leads to an uptick in Romance videos being consumed)
2.  One-off, short term events (e.g. Coronavirus or other disasters, leading to short-term interest in documentaries about them)


<img src="https://miro.medium.com/max/1400/1*M_kpV784Z8fyCUjBvmB7cw.png" alt="drawing" width="1000"/>

## 4.Continue Watching Ranker
This algorithm looks at items that the member has consumed but has not completed, typically:
1. Episodic content (e.g. drama series)
2. Non-episodic content that can be consumed in small bites (e.g. movies that are half-completed, series that are episode independent such as Black Mirror)</br>

The algorithm calculates the probability of the member continue watching and includes other context-aware signals (e.g. time elapsed since viewing, point of abandonment, device watched on, etc).

<img src="https://miro.medium.com/max/1400/1*2YtGx5jWHAhOIC94XMNxfg.png" alt="drawing" width="1000"/>

## 5.Video-Video Similarity Ranker —a.k.a. Because you watched (BYW)
This algorithm basically resembles that of a **content-based filtering algorithm**. Based on an item consumed by the member, the algorithm computes other similar items (using an **item-item similarity matrix**) and returns the most similar items. Amongst the other algorithms, this one is unpersonalised as no other side features are utilised. However, it is personalised in the sense that it is a conscious choice to display a particular item’s similar items a member’s homepage (more details in Page Generation below).
